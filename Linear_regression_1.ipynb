```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('/Users/dyadyundanil/Desktop/Machine learning/DATA/Advertising.csv')
```

```python
df.head()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,4), dpi=200)
fig.suptitle('Scatters for between all features and target')

sns.regplot(data=df, x='TV', y='sales', ax=axes[0])
sns.regplot(data=df, x='radio', y='sales', ax=axes[1])
sns.regplot(data=df, x='newspaper', y='sales', ax=axes[2])
plt.tight_layout()
```

```python
coeffs_1 = np.polyfit(df['TV'], df['sales'], deg=1)
coeffs_2 = np.polyfit(df['radio'], df['sales'], deg=1)
coeffs_3 = np.polyfit(df['newspaper'], df['sales'], deg=1)

coeffs_1_pol = np.polyfit(df['TV'], df['sales'], deg=3)
coeffs_2_pol = np.polyfit(df['radio'], df['sales'], deg=3)
coeffs_3_pol = np.polyfit(df['newspaper'], df['sales'], deg=3)

sales_of_TV_pred = np.polyval(coeffs_1, df['TV'])
sales_of_radio_pred = np.polyval(coeffs_2, df['radio'])
sales_of_newspaper_pred = np.polyval(coeffs_3, df['newspaper'])

sales_of_TV_pred_1 = np.polyval(coeffs_1_pol, df['TV'])
sales_of_radio_pred_1 = np.polyval(coeffs_2_pol, df['radio'])
sales_of_newspaper_pred_1 = np.polyval(coeffs_3_pol, df['newspaper'])
#sales = 0.047 + 7.03*TV
#sales = 0.2 + 9.31*radio
#sales = 0.05 + 12.35*newspaper
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12,6), dpi=200)
fig.suptitle('Scatters for between all features and target')

sns.scatterplot(data=df, x='TV', y='sales', ax=axes[0][0])
sns.lineplot(x=df['TV'], y=sales_of_TV_pred, ax=axes[0][0], color='red')

sns.scatterplot(data=df, x='radio', y='sales', ax=axes[0][1])
sns.lineplot(x=df['radio'], y=sales_of_radio_pred, ax=axes[0][1], color='red')

sns.scatterplot(data=df, x='newspaper', y='sales', ax=axes[0][2])
sns.lineplot(x=df['newspaper'], y=sales_of_newspaper_pred, ax=axes[0][2], color='red')

sns.scatterplot(data=df, x='TV', y='sales', ax=axes[1][0])
sns.lineplot(x=df['TV'], y=sales_of_TV_pred_1, ax=axes[1][0], color='red')

sns.scatterplot(data=df, x='radio', y='sales', ax=axes[1][1])
sns.lineplot(x=df['radio'], y=sales_of_radio_pred_1, ax=axes[1][1], color='red')

sns.scatterplot(data=df, x='newspaper', y='sales', ax=axes[1][2])
sns.lineplot(x=df['newspaper'], y=sales_of_newspaper_pred_1, ax=axes[1][2], color='red')

plt.tight_layout()
```
